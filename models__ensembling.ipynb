# В данном ноутбуке представлены красткие теоретические сведения по ансамлированию моделей

Суть ансамблирования - объединение прогнозов, полученных разными моделями для получения лучшего прогноза.

Ансамблировать стоит модели, каждая из которых максимально отличается от других и рассматривает разные аспекты реальности

Например, есть притча:

Группа незрячих мудрецов впервые встречает слона и, чтобы понять, что такое слон, ощупывает его. Каждый касается только одной его части, такой как туловище или нога. Затем каждый мудрец описывает свое представление о слоне: «он гибкий, как змея», «он как колонна или ствол древа» и т. д.

Вот каждая модель должна рассматривать часть реальности, и все вместе они могут дать отличное полное описание реальности.

Сила - в разнообразии

Например, Ф.Шолее говорит, что в 2014 году они заняли 4 место на Каггле, когда использовали ансамбль из моделей, одна из которых имела очень малую точность. Однако в ансамбле она резко повысила общую точность, потому что запоминала те признаки, которые другие модели не замечали.

```
# самый простой способ ансамблирования - получить среднее прогнозов
pred_a = model_a.predict(x)
pred_b = model_b.predict(x)
pred_c = model_c.predict(x)
pred_d = model_d.predict(x)

final_pred = (pred_a + pred_b + pred_c + pred_d) / 4

# более эффективный способ - вычисление взвешенного среднего
# веса 0.5, 0.25... вычисляются отдельно с помощью методов оптимизации (их не рассматриваем)
final_pred = 0.5 * pred_a + 0.25 * pred_b + 0.1 * pred_c + 0.15 * pred_d
```
